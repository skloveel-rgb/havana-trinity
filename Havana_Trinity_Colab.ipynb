### 3단계: 웹 서버 실행 및 외부 접속 URL 생성 (Localtunnel 터널링 방식)
* 아래 셀을 실행하면 **외부에서 바로 접속할 수 있는 `https://....loca.lt` URL**이 생성됩니다.
* 비밀번호(IP 주소)를 묻는 창이 나오면 아래에 함께 출력되는 **Tunnel Password IP**를 입력하시면 바로 접속됩니다.

```python
import subprocess
import time
import urllib.request

# 1. 백그라운드에서 FastAPI 웹 서버 시작
server_process = subprocess.Popen(["python", "-m", "uvicorn", "app:app", "--host", "0.0.0.0", "--port", "8000"])
time.sleep(4)

# 2. 내 코랩 공개 IP (비밀번호용) 확인
try:
    ip = urllib.request.urlopen('https://ipv4.icanhazip.com').read().decode('utf8').strip()
    print(f'🔑 Tunnel Password (접속 시 입력할 IP): {ip}\n')
except Exception as e:
    pass

# 3. Localtunnel로 외부 접속 URL 생성
print('🌐 생성된 웹 사이트 접속 URL 아래 클릭:')
!npx localtunnel --port 8000
```
